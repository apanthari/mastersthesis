```python
import pandas as pd
import numpy as np
final_df = pd.read_csv('all_features.csv')
```

```python
final_df = final_df.drop(columns=['PubChemCID', 'sixwords', 'notes', 'smiles1',	'category', 'Unnamed: 6'])
final_df.head()
```

### Remove columns that have no values in any row

```python
#split dataframe into tags and substances
tags = final_df.loc[:, '2CB impurity':'weak opioid']
tags.head()
```

```python
substances = final_df.loc[:, 'fentanyl_y':'lysergic acid amide']
substances.head()
```

```python
#lets start with removing the unnecessary substance cols
remove_substance_cols = []
for index in substances.columns:
    
    search_value = 0
    mask = substances[index] == search_value
    result = substances[mask] #this is where there are zeros

    #result are the rows that HAVE a 1 
    if len(result) >= 210: #will show columns that did NOT have a 1
        remove_substance_cols.append(index)
print('number of substances being removed: ', len(remove_substance_cols))
```

```python
#lets start with removing the unnecessary substance cols
remove_tags_cols = []
for index in tags.columns:
    
    search_value = 1
    mask = tags[index] == search_value
    result = tags[mask] #this is where there are zeros

    #result are the rows that HAVE a 1 
    if len(result) <= 5: #will show columns that did NOT have a 1
        remove_tags_cols.append(index)
print('number of tags being removed: ', len(remove_tags_cols))
```
