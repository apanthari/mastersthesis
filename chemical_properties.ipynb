```python
import pubchempy as pcp
import time
import pandas as pd
import numpy as np
```

```python
co_occurences = pd.read_csv('co_occurrences.csv')
co_occurences.head()
```

```python
substance = pcp.Compound.from_cid(165361679)
print(substance.xlogp)
```

```python
final_frame = co_occurences.dropna(subset=['PubChemCID'])
final_frame['PubChemCID'] = final_frame['PubChemCID'].astype(int)
final_frame
```

```python
final_frame = final_frame[final_frame['PubChemCID'] != 485302082]
final_frame
```

```python
for index, row in final_frame[0:50].iterrows():
    CID = row['PubChemCID']
    substance = pcp.Compound.from_cid(CID)
    try:
        weight = substance.molecular_weight
        xlogp = substance.xlogp
        h_donor = substance.h_bond_donor_count
        h_acceptor = substance.h_bond_acceptor_count
        psa = substance.tpsa
    except Exception as e:
        xlogp = None
        h_donor = None
        h_acceptor = None
        psa = None
        
    time.sleep(0.2)  # Add delay to avoid PubChem rate limits

    final_frame.loc[index, 'molecular_weight'] = weight
    final_frame.loc[index, 'XLogP'] = xlogp
    final_frame.loc[index, 'HBondDonorCount'] = h_donor
    final_frame.loc[index, 'HBondAcceptorCount'] = h_donor
    final_frame.loc[index, 'tpsa'] = psa
```

```python
final_frame.to_csv('all_features.csv', index=False)  # index=False prevents writing the index
```
