```python
import numpy as np
from stepmix import StepMix

df = reduced.loc[:, 'adulterant':'weak opioid']
X= df.to_numpy()

model = StepMix(n_components=5, measurement='binary', verbose=1)
model.fit(X)

# Predict latent class memberships
labels = model.predict(X)


# Get class probabilities
probs = model.predict_proba(X)

# Print results
print("Labels:", labels)
print("Probabilities:\n", probs)


# Model parameters
print("Parameters:\n", model.get_parameters())
```

```python
# Get model parameters using get_parameters or get_parameters_df
parameters = model.get_parameters()

# Access the pis (item parameters) from the measurement section
item_params = parameters['measurement']['pis']

n_classes, n_items = item_params.shape

# Create item labels (you can customize these based on your actual items)
item_labels = [f"Item {i+1}" for i in range(n_items)]

# Create class labels (you can customize these based on your actual latent classes)
class_labels = [f"Class {i+1}" for i in range(n_classes)]

# Convert the item_params array into a pandas DataFrame
item_response_table = pd.DataFrame(item_params, columns=item_labels, index=class_labels)

# Display the item response table
item_response_table
```

```python
# Step 1: Get conditional probabilities
emission_probs = model.get_parameters()["measurement"]["pis"]  # shape: (n_classes, n_features)

# Step 2: Compute variability across classes
feature_importance_scores = np.std(emission_probs, axis=0)

# Step 3: Get actual column names
feature_names = df.columns.tolist()  # works if X is a DataFrame

# Step 4: Rank features by importance
importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Importance": feature_importance_scores
}).sort_values(by="Importance", ascending=False)

importance_df.head(17)
```

```python
# If not already done:
n_classes, n_features = emission_probs.shape

# Build a DataFrame for each class
class_tables = []
for class_idx in range(n_classes):
    probs = emission_probs[class_idx, :]  # shape: (n_features,)
    df = pd.DataFrame({
        "Feature": feature_names,
        f"Prob_Class_{class_idx}": probs
    }).sort_values(by=f"Prob_Class_{class_idx}", ascending=False)
    class_tables.append(df)
```

```python
top_features_per_class = {}

for class_idx, df in enumerate(class_tables):
    top_features = df.head(10)
    top_features_per_class[f"Class_{class_idx}"] = list(top_features["Feature"])

# View the summary
df = pd.DataFrame(top_features_per_class)
df
```
