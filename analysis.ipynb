```python
#figure out if its an adulterant or not
adulterants.loc[(adulterants['expectedsubstance'] == adulterants['substance']), 'adulterant'] = 0 #not an adulterant
adulterants.loc[(adulterants['expectedsubstance'] != adulterants['substance']), 'adulterant'] = 1 #an adulterant
```

```python
adulterants
```

```python
#get the mean value of adulterant column grouped by sample id
adulterants['adulterant_percentage'] = adulterants.groupby('sampleid')['adulterant'].transform(lambda x: x.mean())
adulterants
```
