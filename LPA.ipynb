```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
from stepmix import StepMix
```

```python
reduced = pd.read_csv('dimension_reduction.csv')
reduced
```

```python
substances = reduced.loc[:, 'fentanyl_y':]
substances
```

```python
X= substances.to_numpy()
X
```

```python
# Optional: Standardize the variables (important for Gaussian assumptions)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# StepMix model setup for continuous (Gaussian) indicators
model = StepMix(n_components=3,  # change based on expected number of latent profiles
                n_steps=1,
                measurement='continuous',  # key change for LPA
                verbose=1,
                random_state=42)

# Fit the model
model.fit(X_scaled)

# Get predicted class assignments
latent_classes = model.predict(X_scaled)

# Class probabilities for each individual
class_probs = model.predict_proba(X_scaled)

# Means for each latent class
class_means = model.get_parameters()["means_"]

print("Predicted Latent Classes:", latent_classes)
print("Class Probabilities:", class_probs)
print("Class Means:\n", class_means)
```
